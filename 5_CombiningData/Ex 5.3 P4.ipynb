```python
from IPython.display import display, HTML
import pandas as pd
import math

import plotly.graph_objects as go
import plotly.express as px
import numpy as np
from scipy import special
```

<h3> State Data </h3>

```python
#Read data
df_covid = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv', dtype={"fips": str})

print("Number of Rows:  ", df_covid.shape[0])
print("Number of Columns:  ", df_covid.shape[1])
df_covid.head()
```

```python
# NC row filter
row_filter = df_covid['state'] == 'North Carolina'

df_NC = df_covid[row_filter]

df_NC.head()
```

```python
# Plot

fig = px.line(df_NC,
             x='date',
             y='cases',
             template='presentation',
             title='Question 3: How has the number of COVID-19 Cases in NC grown?')

fig.show()
```

<h3>County Data</h3>


```python
#Read data
df_cv_county = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv', dtype={"fips": str})

print("Number of Rows:  ", df_cv_county.shape[0])
print("Number of Columns:  ", df_cv_county.shape[1])
df_cv_county.head()
```

```python
# NC row filter
row_filter = df_cv_county['state'] == 'North Carolina'

df_NC = df_cv_county[row_filter]

df_NC.head()
```

```python
# Plot

fig = px.line(df_NC,
             x='date',
             y='cases',
             template='presentation',
              color='county',
             title='Question 4: How has the number of COVID-19 Cases in NC grown - by County?')

fig.show()
```

```python
# County filter

NC_counties = ['Wake','Mecklenburg','Buncombe','Jackson']
```

```python
row_filter = df_NC['county'].isin(NC_counties)

df_NC_counties = df_NC[row_filter]

df_NC_counties.head()
```

```python
# Plot

fig = px.line(df_NC_counties,
             x='date',
             y='cases',
             template='presentation',
              color='county',
             title='Question 4: How has the number of COVID-19 Cases in NC grown - by County?')

fig.show()
```
