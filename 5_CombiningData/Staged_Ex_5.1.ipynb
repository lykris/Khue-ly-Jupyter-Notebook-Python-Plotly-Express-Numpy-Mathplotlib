# In Class: Append September data to the July/Aug data  
1. Load prior Appended data from csv file named *mnth_current.csv* into a Dataframe named: *df_old_current*    
2. Read new csv data (to append) into a Dataframe named: *df_mnth9_2011.csv*  
3. Append df_mnth9_2011 to df_old_current into a new dataframe named: *df_new_current*  
4. Save the appended dataframe by overwriting the existing *mnth_current.csv* with the new data  

### 1. Load prior Appended data csv file into a Dataframe
Should have:  
- Number of Rows:   22  
- Number of Columns:   6

```python
df_old_current = pd.read_csv("Data/mnth_current.csv")

# Display of number of rows and columns in the dataframe to doublecheck
print("Number of Rows:  ", df_old_current.shape[0])
print("Number of Columns:  ", df_old_current.shape[1])
df_old_current.head()
```

###  2. Read new csv data (to append) into a Dataframe    
- Filename:  mnth9_2011.csv
- Number of Rows:   10
- Number of Columns:   6

```python
#Read the mnth9_2011.csv file into a pandas dataframe
df_mnth9_2011 = pd.read_csv("Data/mnth9_2011.csv")

print("Number of Rows:  ", df_mnth9_2011.shape[0])
print("Number of Columns:  ", df_mnth9_2011.shape[1])
```

```python
#Display the first three rows in the dataframe
df_mnth9_2011.head(3)
```

### 3. Append df_mnth9_2011 to df_old_current 
After appending should end up with:  
- Number of Rows:   32  
- Number of Columns:   6

```python
# Append df_mnth9_2011 to df_Appended_8_2011
df_new_current = df_old_current.append(df_mnth9_2011, ignore_index=True)
```

```python
# Append check for new df_new_current
print("Number of Rows:  ", df_new_current.shape[0])
print("Number of Columns:  ", df_new_current.shape[1])
df_new_current.head(2)
```

### 4. Overwrite the existing *mnth_current.csv* file with this newly appended data  
- After saving the csv file, be sure to test it by reading it back into a dataframe and verifying the number of rows and columns are correct.

```python
# Overwrite the existing csv file
df_new_current.to_csv("Data/mnth_current.csv", index=False)
```

```python
# Test

df_test = pd.read_csv("Data/mnth_current.csv")

print("Number of Rows:  ", df_test.shape[0])
print("Number of Columns:  ", df_test.shape[1])
df_test.head(2)
```
